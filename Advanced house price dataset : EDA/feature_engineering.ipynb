## Advanced Housing Prices- Feature Engineering
The main aim of this project is to predict the house price based on various features which we will discuss as we go ahead

Dataset to downloaded from the below link
https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

We will be performing all the below steps in Feature Engineering

1. Missing values
2. Temporal variables
3. Categorical variables: remove rare labels
4. Standarise the values of the variables to the same range

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns",None)
```

```python
dataset=pd.read_csv("train.csv")
dataset=dataset.drop(columns=['Id'])
dataset.head()
```

### **categorical features null handling**

```python
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=="O"]
```

```python
feature_nan
```

```python
def filling_missing_val(dataset,feature_nan):
    
    dataset[feature_nan]=dataset[feature_nan].fillna("missing")
    return dataset
dataset=filling_missing_val(dataset,feature_nan)
dataset[feature_nan].isnull().sum()
```

### **numerical with null handling**

```python
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
numerical_with_nan
```

```python
for feature in numerical_with_nan:
    print("{} is {}% of missing values".format(feature,np.round(dataset[feature].isnull().mean()*100,2)))
```

### **feature engineering to add columns with 1 for null and 0 for non null**

```python
# replacing numerical missing values 

for feature in numerical_with_nan:
    median_val=dataset[feature].median()
    dataset[feature +" "+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_val,inplace=True)
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.shape
```

### **handle temp year features to feature engineering to get  column as ["age_of_house","years_remod","years_to_garagebuilt"]**

```python
for feature in ["YearBuilt","YearRemodAdd","GarageYrBlt"]:
    dataset[feature+"age"]=dataset["YrSold"]-dataset[feature]
    
```

```python
dataset[["YearBuiltage","YearRemodAddage","GarageYrBltage"]].head()
```

```python
dataset.rename(columns={"YearBuiltage":"age_of_house","YearRemodAddage":"years_to_remod","GarageYrBltage":"years_to_garagebuilt"},inplace=True)
```

```python
dataset.head()
```

### **separate each features**

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
year_column=[feature for feature in numerical_features if "Yr"  in feature or "Year" in feature]
```

```python
year_column=[col for col in year_column if col!='GarageYrBlt nan']
```

```python
year_column
```

```python
continuous_feature=[feature for feature in numerical_features if feature not in year_column and len(dataset[feature].unique())>25]
```

```python
for feature in continuous_feature:
    if 0 in dataset[feature].unique():
        pass
    else:
        dataset[feature]=np.log(dataset[feature])
        
```

```python
dataset.head()
```

### **handling rare categorical features**

```python
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
categorical_feature
```

```python
for feature in categorical_feature:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'rare_var')
```

```python
dataset.head(89)
```

```python
numerical_features
```

```python
for feature in categorical_feature:
    labels_ordered=dataset.groupby(feature)['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head(100)
```

### **feature scaling**

```python
scaling_feature=[feature for feature in dataset.columns if feature not  in['SalePrice']]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit_transform(dataset[scaling_feature])
```

```python
data = pd.concat([dataset[['SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[scaling_feature]), columns=scaling_feature)],
                    axis=1)
```

```python
data.to_csv("X_train.csv",index=False)
```
